
### INTRODUCCION:
El conjunto de datos contiene información sobre propiedades inmobiliarias, específicamente en relación con la **antigüedad de la construcción** y el **precio de las casas**. La **antigüedad** se mide en años, mientras que el **precio** está expresado en pesos, ajustado a una escala de 1:100,000 para facilitar su visualización y análisis.

El objetivo principal del uso del algoritmo **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** es identificar agrupaciones o clusters en los datos que compartan características similares en cuanto a antigüedad y precio. A través de este enfoque, buscamos segmentar las viviendas en grupos con patrones comunes, lo que nos permitirá reconocer diferentes tipos de propiedades, como aquellas con precios más bajos y antiguas, frente a las más nuevas y caras.

Con DBSCAN, también pretendemos identificar puntos atípicos o **ruido**, es decir, propiedades que no se ajustan claramente a ningún grupo. Estos puntos pueden representar casos excepcionales o datos que se desvían significativamente de las tendencias generales del mercado.

En resumen, al aplicar DBSCAN, buscamos entender mejor la distribución de precios y antigüedad en el mercado inmobiliario, identificar agrupaciones relevantes para el análisis y segmentar las propiedades de manera efectiva, lo que podría ser útil para estrategias de marketing, inversión y análisis de tendencias en bienes raíces.<BR>

### 1. Importamos las librerias necesarias:

```python
# Importamos las libreriras
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```


<ul>
<li> Con pandas manipularemos el dataframe.
<li> Con matplotlib realizarmos el grafico.
<li> Importaremos el módulo DBSCAN para ejecutar el algoritmo que agrupe la variables.
<ul>

### 2. Cargamos el dataset

```python
# Cargamos el dataset
# ID del archivo de Google Drive
file_id = '1WfIOq8JJp38f0QjX1zOBiT7blLtySf0E'

# Generar URL directa y leer el CSV
url = f'https://drive.google.com/uc?id={file_id}'
df = pd.read_csv(url)
```

### 3. Identificamos los clusters

```python
# Identificamos los clusters
clusters = DBSCAN(eps=2, min_samples=10).fit_predict(df)
print(clusters)
```

El código utiliza el algoritmo DBSCAN (Density-Based Spatial Clustering of Applications with Noise) para identificar los clusters en un conjunto de datos.<br>
<ol>
<li>DBSCAN;
<ul><li>DBSCAN es un algoritmo de clustering basado en la densidad que agrupa puntos que están cercanos entre sí y marca como ruido aquellos puntos que no cumplen con las condiciones de densidad.
<li>A diferencia de otros algoritmos como K-means, DBSCAN no requiere que se especifique el número de clusters previamente, lo que lo hace muy útil cuando no se sabe cuántos clusters esperar.
</ul>
<li>eps=2;
<ul>
<li>eps es el parámetro de distancia máxima que se usa para considerar que dos puntos están en el mismo grupo. Es decir, si la distancia entre dos puntos es menor o igual a eps, se consideran parte del mismo cluster.
<li>En este caso, eps=2 significa que cualquier par de puntos cuya distancia euclidiana sea menor o igual a 2 serán considerados parte del mismo cluster.
<li>Si eps es muy pequeño, puede haber muchos puntos considerados como ruido (outliers). Si es muy grande, puede hacer que los puntos de diferentes grupos sean agrupados juntos.
</ul>
<li>min_samples=10:
<ul>
<li>min_samples es el número mínimo de puntos requeridos para formar un cluster. Si un punto tiene al menos min_samples puntos dentro del rango de eps, entonces ese punto será considerado un punto central de un cluster, y los puntos cercanos a él se considerarán parte del mismo cluster.
<li>    En este caso, min_samples=10 significa que para que un grupo de puntos sea considerado un cluster, debe tener al menos 10 puntos en un radio de eps=2.
</ul>

<li>.fit_predict(df):
<ul>
<li>fit_predict es un método de DBSCAN que ajusta el modelo de clustering a los datos (realiza el fit) y luego predice a qué cluster pertenece cada punto.
<li>El argumento df es el DataFrame con los datos de entrada, que deben contener las características relevantes sobre las que se va a hacer el clustering.
<li>El resultado de fit_predict(df) es un array que contiene los identificadores de los clusters para cada punto. Los puntos que no pertenecen a ningún cluster (puntos de ruido) reciben un valor de -1.
</ul>
</ol>
<br>

###**Resultado:**
<ul>
clusters será un array de números enteros, donde cada número representa el cluster asignado a cada punto del DataFrame df. Los puntos de ruido (que no cumplen con las condiciones de densidad) tendrán el valor -1.
</ul>

### 4. Graficamos y analizamos los clusters:

```python
# Graficamos con matplotlib y mostramos los clusters
plt.figure(figsize=(6, 6))

plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters, s=80)

plt.xlabel("Antigüedad de la construcción en años")
plt.ylabel("Precio (1:100.000)")

plt.show()
```

Clusters distintos: El gráfico muestra claramente tres grupos (clusters) con colores distintos, lo que sugiere que el algoritmo de clustering (probablemente DBSCAN u otro) ha identificado tres regiones bien definidas en el espacio de las variables "Antigüedad de la Construcción en Años" y "Precio de Casa".
